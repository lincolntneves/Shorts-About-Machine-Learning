# qcut() for build a scores to the classification Machine Learning Model

The objective of this notebook is only to show up how to use pandas.qcut() to build a score classes to the classification ML model

```python
# Loading de necessary packages
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from scipy import stats
```

```python
# Loading the dataset Iris from sklearn
load_breast_cancer = load_breast_cancer()
```

```python
# Organizing X and y to train a machine learning model
X = load_breast_cancer['data']
y = load_breast_cancer['target']
```

```python
# Building my machine learning model with the train subset
# Using 'probability=True' because I'll need to use on 'predict_proba'
clf = svm.SVC(probability=True)
clf.fit(X, y)
```

```python
# Here I'm using 'preditct' to have the flag 0 or 1, but I'm also using
# 'predict_proba' to have the probability of the event happens
y_valid = clf.predict(X)
y_valid_proba = clf.predict_proba(X)
```

```python
# I need to transform the array to datafram to inser the new columns with qcut()
df = pd.DataFrame(np.c_[load_breast_cancer['data'], load_breast_cancer['target']],
                  columns= np.append(load_breast_cancer['feature_names'], ['target']))
df.head()
```

```python
# Now I'm puting on the dataframe three new columns
# On the first and second columns have the probability the target and no target
# On the other one I have the predict class with 1 or 0
df['predict_0'] = y_valid_proba[:,0]
df['predict_1'] = y_valid_proba[:,1]
df['predict_class'] = y_valid
df.head()
```

```python
# Here I'm building 10 classes about probability from predict_proba() and after 
# I'm appending a score from 1 to 10 for eath classe
df['cut_prob_10'] = pd.qcut(df.predict_1, 10)
df['cut_10'] = pd.qcut(df.predict_1, 10, labels=[i for i in range(1,11)])
```
