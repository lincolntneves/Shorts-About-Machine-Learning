# Predict and Predict Proba for Classification Machine Learning Model

The objective of this notebook is only to show up how to use Predict, that results on a dummy variable, and Predict Proba that results on a probability to happens the event

```python
# Loading de necessary packages
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from scipy import stats
```

```python
# Loading the dataset Iris from sklearn
load_breast_cancer = load_breast_cancer()
```

```python
# Organizing X and y to train a machine learning model
X = load_breast_cancer['data']
y = load_breast_cancer['target']
```

```python
# Split your dataframe into random train and test subsets
# On this case I use 30% of the dataframe to test and 70% to train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Building my machine learning model with the train subset
# Using 'probability=True' because I'll need to use on 'predict_proba'
clf = svm.SVC(probability=True)
clf.fit(X_train, y_train)
```

```python
# Here I'm using 'preditct' to have the flag 0 or 1, but I'm also using
# 'predict_proba' to have the probability of the event happens
y_valid = clf.predict(X_test)
y_valid_proba = clf.predict_proba(X_test)
```

```python
# An array with 1 or 0, depends of the result of probability to happen
y_valid
```

```python
# An array with the probability to the event happens (1) or not heppens (0)
# The sum of both is 100%
y_valid_proba
```
